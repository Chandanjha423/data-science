```python
import sys
import matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')


X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])

plt.bar(X, Y, color='g')

plt.title("Students over 11 Years")

plt.xlabel("Years")

plt.ylabel("Number of Students")
plt.plot()

plt.show()
```
