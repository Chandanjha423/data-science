```python
# pandas series-: a pandas series is like a column in table. it is one_dimensional array holding data of any type
a =[1,7,2]
x = pd.Series(a)
print(x)
```

```python
# labels-:if nothing else is specified the value are labeled with there index number first value are labeled with there index number 
a=[1, 2, 7]
x=pd.Series(a)
print(x[0])
```

```python
#create labels:-with the index argument you can name your own labels 
import pandas as pd 
a=[1, 2, 7]
x=pd.Series(a,index=["x","y","z"])
print(x)
```

```python
import pandas as pd
a=[1,7,2]
x=pd.Series(a,index=["x","y","z"])
print(x["y"])
```
